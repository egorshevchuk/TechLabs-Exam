```python
df = df.drop(['host_id'], axis = 1)
df
```

D) Only show listings with a price higher than the average price. Sort the dataframe by price in an ascending order.

```python
df_sorted = df.loc[df['price'] > df['price'].mean()]
df_sorted.sort_values(by = ['price'], inplace = True)
df_sorted
```

E) Only show listings that have more than the average reviews and that are available for more than 10 nights (column "minimum_nights"). Sort the dataframe by reviews in a descending order.

```python
df_sorted_2 = df.loc[df['number_of_reviews'] > df['number_of_reviews'].mean()]
df_sorted_2 = df.loc[df['minimum_nights'] > 10]
df_sorted_2.sort_values(by = ['number_of_reviews'], ascending = False, inplace = True)
df_sorted_2
```

J) Scale the column "price" and the column "number_of_reviews" using a method that seems appropriate in your opinion. </p>
Save the new dataframe again under the name "df".

```python
# First, to decide on normalization method we need to see how the data looks like and check descriptive statistics
# 1. Price
sns.displot(df, x = 'price')
df['price'].describe()
# 2. Number of reviews
sns.displot(df, x = 'number_of_reviews')
df['number_of_reviews'].describe()

# For price the distibution has a positive scew. Plus, it has several extreme outliers
# Thus, we should first use feature clipping (assign all values above some threshold to this threshold and all zero-like values)
# Then we will proceed with log-scaling
df['price'].clip(lower = 10, upper = 1000, inplace = True)
df['price_log'] = np.log(df['price'])
sns.displot(df, x = 'price_log')

# For number of reviews the distribution is equivalent
df['number_of_reviews'].clip(upper = 200, inplace = True)
df['number_of_reviews_log'] = np.log(df['number_of_reviews'])
sns.displot(df, x = 'number_of_reviews_log')
# Here, the result is not ideal - too many variables in the lower part of the distribution. However, other methods will give
# even worse results
```
